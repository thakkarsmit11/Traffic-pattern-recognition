```python
Dist = pd.read_csv(r"C:\Users\n10387684\OneDrive - Queensland University of Technology\1_QUT\Codes\GitHub\TTR_AS\Data_Finucane road\Proposed_Distance_Matrix.csv",header = 0)
dist = Dist.drop(Dist.iloc[:, 0:1], axis=1)
```

```python
square = squareform(dist) # sqareform distance matrix for dendogram viz.
Z = linkage(square, 'ward')
```

```python
fig = plt.figure(figsize=(150, 100))
ax = fig.add_subplot(1, 1, 1)

dn = dendrogram(Z,  p=8, leaf_font_size = 150, leaf_rotation=90, color_threshold=0.72, above_threshold_color='blue', ax=ax)
ax.tick_params(axis='y', which='major', labelsize=200)

# Cutting the dendrogram at max_d
plt.axhline(y=0.72, c='k',linestyle='--')

plt.rcParams["font.weight"] = "bold"
plt.rcParams["axes.labelweight"] = "bold"

plt.rcParams['axes.linewidth'] = 4
plt.rcParams['lines.linewidth'] = 15

list_labels = [item.get_text() for item in ax.get_xticklabels()]
labels = list(map(int, list_labels))
new_labels = [x+1 for x in labels]
str_new_labels = list(map(str, new_labels))

ax.set_xticklabels(str_new_labels)
ax.set_xticks(ax.get_xticks()[::5])

plt.ylabel("Composite distance", fontsize=250)
plt.xlabel("Time series", fontsize=250)
plt.subplots_adjust(left=0.1, right=0.11, top=0.11, bottom=0.1)



fig.tight_layout()
plt.show()
```
